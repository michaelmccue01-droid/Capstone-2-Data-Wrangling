## Data Collection

The dataset used in this project is the Zillow Home Value Index (ZHVI), downloaded from Kaggle.

It contains monthly home value estimates for U.S. states from 2000â€“2026.

The raw file was stored in the data/raw directory and loaded into the notebook using pandas.

```python
df.index.name = "date"

df_long = df.reset_index().melt(
    id_vars="date",
    var_name="state",
    value_name="zhvi"
)
```

```python
df_long['state'].nunique()
df_long['date'].min(), df_long['date'].max()
```

```python
df_long.shape
```

```python
df_long.head()
```

```python
df_long.duplicated(subset=['date', 'state']).sum()
```

```python
df_long.isna().sum()
```

```python
df_long[df_long['zhvi'].isna()].head()
```

```python
# Gives the percentage missing
df_long['zhvi'].isna().mean() * 100
```

```python
df_long = df_long.dropna(subset=['zhvi'])
df_long.isna().sum()
```

```python
df_long['zhvi'].describe()
```

```python
df_long['state'].nunique()
```

```python
df_long = df_long.sort_values(['state', 'date'])
```

```python
import os

os.makedirs('../data/processed', exist_ok=True)
df_long.to_csv('../data/processed/zhvi_long_clean.csv', index=False)
```
